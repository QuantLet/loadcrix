```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("white")
    
```

```python
cutoff    = '2016-05-02'

crix      = pd.read_csv('crix.csv')#, skiprows=1)
crix      = crix[crix['date'] >= cutoff] 
crix.date = pd.to_datetime(crix.date)
crix.set_index('date', inplace=True)
crix.plot(figsize=(15,5), linewidth=5, fontsize=20, color= 'darkgrey',legend=False)
plt.xlabel('Year', fontsize=20);
plt.savefig(f'.TS_CRIX.png', transparent=True)

vcrix     = pd.read_csv('vcrix.csv')#, skiprows=1)
vcrix     = vcrix[['date','vcrix']]
vcrix      = vcrix[vcrix['date'] >= cutoff] 
vcrix.date = pd.to_datetime(vcrix.date)


vcrix.set_index('date', inplace=True)
vcrix.plot(figsize=(15,5), linewidth=5, fontsize=20, color= 'orange',legend=False)
plt.xlabel('Year', fontsize=20);
plt.savefig(f'.TS_VCRIX.png', transparent=True)
```
